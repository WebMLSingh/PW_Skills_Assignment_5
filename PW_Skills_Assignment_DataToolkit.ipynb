14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
* a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
* b) Label the x-axis as 'Date'.
* c) Set the title of the plot as 'Temperature and Humidity Over Time'.


```python
# Generate a time-series dataset
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.uniform(low=15, high=35, size=100)  # Random temperature values between 15 and 35 degrees
humidity = np.random.uniform(low=40, high=90, size=100)      # Random humidity values between 40% and 90%

# Create a DataFrame
data = {'Date': dates, 'Temperature': temperature, 'Humidity': humidity}
df = pd.DataFrame(data)

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], 'r-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (Â°C)', color='r')
ax1.tick_params('y', colors='r')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], 'b-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='b')
ax2.tick_params('y', colors='b')

# Set title and format x-axis for dates
plt.title('Temperature and Humidity Over Time')
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
* a) Plot a histogram of the data with 30 bins.
* b) Overlay a line plot representing the normal distribution's probability density function (PDF).
* c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
* d) Set the title of the plot as 'Histogram with PDF Overlay'

```python
# Generate 1000 samples from a normal distribution
data = np.random.randn(1000)

# Create the histogram
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Calculate and plot the PDF (Probability Density Function)
mu, sigma = 0, 1  # Mean and standard deviation for standard normal distribution
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((bins - mu) / sigma) ** 2)
plt.plot(bins, pdf, 'r-', lw=2, label='PDF')

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Display legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import seaborn as sns

# Create two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Determine quadrants
quadrants = []
for i in range(len(x)):
    if x[i] >= 0 and y[i] >= 0:
        quadrants.append('Quadrant I')
    elif x[i] < 0 and y[i] >= 0:
        quadrants.append('Quadrant II')
    elif x[i] < 0 and y[i] < 0:
        quadrants.append('Quadrant III')
    else:
        quadrants.append('Quadrant IV')

# Create a DataFrame for Seaborn
df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Create scatter plot with coloring based on quadrants
plt.figure(figsize=(8, 6))
sns.scatterplot(x='x', y='y', hue='Quadrant', data=df, palette='deep')

# Label axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set title
plt.title('Quadrant-wise Scatter Plot')

# Add a legend
plt.legend(title='Quadrants')

# Show the plot
plt.tight_layout()
plt.show()
```

18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'

```python
from bokeh.plotting import figure, show
# Generate data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a Bokeh plot
p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis')

# Plot the sine wave
p.line(x, y, line_width=2)

# Add grid lines
p.xgrid.visible = True
p.ygrid.visible = True

# Show the plot
show(p)
```

19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'

```python
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
import random
import string
import pandas as pd

# Step 1: Generate random categorical data
categories = [''.join(random.choices(string.ascii_uppercase, k=5)) for _ in range(10)]
values = [random.randint(10, 100) for _ in range(10)]

# Create a DataFrame to hold the data
data = pd.DataFrame({
    'Category': categories,
    'Value': values
})

# Step 2: Create a ColumnDataSource
source = ColumnDataSource(data)

# Step 3: Create a figure for the bar chart
p = figure(x_range=categories,
           title="Random Categorical Bar Chart",
           toolbar_location=None,
           tools="")

# Step 4: Create the bar chart with color mapping
p.vbar(x='Category',
       top='Value',
       width=0.9,
       source=source,
       line_color='white',
       fill_color=factor_cmap('Category', palette="Viridis256", factors=categories))

# Step 5: Add hover tooltips to display the exact values
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Step 6: Label the axes and rotate x-axis labels for readability
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = "Category"
p.yaxis.axis_label = "Value"
p.xaxis.major_label_orientation = 1.2

# Step 7: Display the bar chart
show(p)

```

20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)  # For reproducibility
x_values = np.arange(10)  # X values (0 to 9)
y_values = np.random.rand(10) * 100  # Random Y values scaled to 0-100

# Create a line plot
fig = go.Figure()

# Add a trace for the line
fig.add_trace(go.Scatter(x=x_values, y=y_values, mode='lines+markers', name='Random Data'))

# Update layout with title and axis labels
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X Axis',
    yaxis_title='Y Axis'
)

# Show the plot
fig.show()
```

21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)  # For reproducibility
labels = ['A', 'B', 'C', 'D', 'E']  # Pie chart labels
values = np.random.randint(1, 100, size=len(labels))  # Random values

# Create a pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values,
                               textinfo='label+percent',
                               hoverinfo='label+percent')])

# Update layout with title
fig.update_layout(title='Interactive Pie Chart')

# Show the plot
fig.show()
```
